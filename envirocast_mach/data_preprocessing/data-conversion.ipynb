## Weather Data

```python
import json

# Load JSON data from file with explicit encoding
with open('../raw_data/data_airquality(april-24).json', 'r', encoding='utf-8') as file:
    data = json.load(file)
```

## Merge Data

## Check Missing Hours

```python
import pandas as pd

# Load CSV file into pandas DataFrame
df = pd.read_csv('data_final.csv')

# Convert 'timestamp_local' column to datetime
df['timestamp_local'] = pd.to_datetime(df['timestamp_local'])

# Sort DataFrame by 'timestamp_local'
df = df.sort_values(by='timestamp_local')
# Reset index after sorting
df = df.reset_index(drop=True)
# Calculate difference between consecutive timestamps
time_diff = df['timestamp_local'].diff()

# Check for missing hours
missing_hours = time_diff[time_diff != pd.Timedelta(hours=1)]

if missing_hours.empty:
    print("No missing hours.")
else:
    print("Missing hours:")
    print(missing_hours)
```

```python
missing_indices = time_diff[time_diff != pd.Timedelta(hours=1)].index

# Display the indices
print("Indices of missing hours:", missing_indices)
```

```python
import pandas as pd

# Load CSV files into pandas DataFrames
df_aqi = pd.read_csv('data_aqi.csv')
df_weather = pd.read_csv('data_weather.csv')

# Extract 'timestamp_local' column from each DataFrame
timestamps_aqi = df_aqi['timestamp_local']
timestamps_weather = df_weather['timestamp_local']

# Find missing indices in df_weather
missing_indices_weather = timestamps_aqi[~timestamps_aqi.isin(timestamps_weather)].index

# Find missing indices in df_aqi
missing_indices_aqi = timestamps_weather[~timestamps_weather.isin(timestamps_aqi)].index

print("Missing indices in data_weather.csv:", missing_indices_weather)
print("Missing indices in data_aqi.csv:", missing_indices_aqi)
```

```python
# Drop duplicates based on 'timestamp_local' column
df_weather = df_weather.drop_duplicates(subset=['timestamp_local'])

# Reset index
df_weather = df_weather.reset_index(drop=True)
```

```python
df_weather.shape
```

```python
# Drop duplicates based on 'timestamp_local' column
df_aqi = df_aqi.drop_duplicates(subset=['timestamp_local'])

# Reset index
df_aqi = df_aqi.reset_index(drop=True)
```

```python
df_aqi.shape
```

```python
df_aqi.to_csv("data_aqi.csv",index=False)
```

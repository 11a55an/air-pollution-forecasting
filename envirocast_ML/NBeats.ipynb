## Loading Data

```python
import pandas as pd
```

```python
df = pd.read_csv("data/data_imputed_2.csv")
```

```python
df.drop(columns=['timestamp_local','city_name','country_code'],inplace=True)
```

```python
temp = df['temp']
```

```python
aqi = df['aqi']
```

```python
co = df['co']
```

```python
no2 = df['no2']
```

```python
so2 = df['so2']
```

```python
pm25 = df['pm25']
```

```python
pm10 = df['pm10']
```

## Temp

```python
from nbeats_tensorflow.model import NBeats
```

```python
# Fit the model
model = NBeats(
    y=temp,
    forecast_period=168,
    lookback_period=400,
    units=30,
    num_blocks_per_stack=2,
    share_weights=True,
    share_coefficients=False,
)
```

```python
model.fit(
    loss='mse',
    epochs=100,
    batch_size=32,
    learning_rate=0.003,
    backcast_loss_weight=0.5,
    verbose=True
)
```

```python
# Generate the forecasts and backcasts
preds_temp = model.forecast(y=temp, return_backcast=False)
```

```python
preds_temp
```

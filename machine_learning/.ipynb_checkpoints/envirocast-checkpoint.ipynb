# EnviroCast - Smart Air Monitoring & Forecasting System

## Import Data

```python
import pandas as pd
```

```python
data = pd.DataFrame(columns=['city_name', 'country_code', 'datetime', 'co', 'no2', 'o3', 'so2', 'pm25', 'pm10', 'aqi', 'timestamp_local', 'timestamp_utc', 'ts'])
```

```python
df = pd.read_json("data_airquality(jan_23).json")
```

```python
df.head()
```

```python
normalized_data = pd.json_normalize(df['data'])
```

```python
normalized_data['city_name'] = "Gujrāt"
```

```python
normalized_data['country_code'] = "PK"
```

```python
normalized_data
```

```python
data = pd.concat([data, normalized_data])
```

```python
data.reset_index(drop=True, inplace=True)
```

```python
data.head()
```

```python
data.info()
```

```python
data['datetime'] = pd.to_datetime(data['datetime'], format='%Y-%m-%d:%H')
```

```python
data = data.sort_values(by='datetime')
```

```python
data.reset_index(drop=True, inplace=True)
```

```python
data.head()
```

```python
data.tail()
```

```python
data.to_csv("data_2023.csv")
```

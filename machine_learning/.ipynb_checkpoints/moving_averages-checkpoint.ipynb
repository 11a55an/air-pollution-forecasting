## Weighted Moving Average

Weighted moving average = (t * weighting factor) + ((t-1) * weighting factor-1) + ((t-n) * weighting factor-n)/n

```python
def wma(weights): 
  def calc(x):
    return (weights*x).mean()
  return calc
```

### NO2

```python
no2_df['no2'].rolling(window=3).apply(wma(np.array([0.5,1,1.5])))
```

```python
no2_df['wma_rolling_3']=no2_df['no2'].rolling(window=3).apply(wma(np.array([0.5,1,1.5]))).shift(1)
```

```python
no2_df.plot()
```
